```python
# analyze smile reponse
smile_data = pd.read_csv('smile_response.csv',header=None)
```

```python
idx = [str(x) for x in range(721)]
idx = ['subject','ad'] + idx
smile_data.columns = idx
smile_data.set_index(['subject','ad'])
print('smile response')
smile_data.ix[:10,:20]
```

```python
# print each group
by_ad = smile_data.groupby('ad')
cnt = 1
fig = figure(figsize=(17,10))

for idx,group in by_ad:
    subplot(2,4,cnt)
    group.mean(axis=0).plot()
    axis([0, 350, 0, .6])
    title(idx)
    xlabel('frame'),ylabel('smile response')
    cnt += 1
print('Overall, some ads results in different level of smile responses')
```

```python
## divide data to zapping and non-zapping
zap, nonzap = [], []
smile_np = array(smile_data)
eps, threshold, cnt = 1e-5, .9, 0
data_num, data_dim = smile_np.shape


for row in smile_np:
    if row[threshold*data_dim] < eps :
        zap.append(row[2:-1])
    else:
        nonzap.append(row[2:-1])
    cnt += 1
zap = np.array(zap)
nonzap = np.array(nonzap)
```

```python
# plot mean smile response for each group

zap_avg = np.mean(zap,axis=0)
nonzap_avg = np.mean(nonzap,axis=0)
plot(zap_avg,label='zapping')
plot(nonzap_avg,label='non-zapping')
title('Average Smile Response')
xlabel('frame')
ylabel('smile response')
legend(loc=2)
axis([0, 720, 0, .5])
```

```python
# plot max smile response

zap_max = np.max(zap,axis=1)
nonzap_max = np.max(nonzap,axis=1)
hist(zap_max, bins=20, alpha=.4, label='zap')
hist(nonzap_max, bins=20, alpha=.4, label='non-zap')
legend(loc=2)
title('Max Smile Response Distribution')
```

```python
# plot the effect of smile volume

N = 100
zap_num, data_dim = zap.shape
nonzap_num = nonzap.shape[0]
print zap_num, nonzap_num
perc, smile_level = np.linspace(0,1,N+1), np.linspace(0,1,N+1) # smile and percentage increment step 0.2
m_zap,m_non = np.zeros((N+1,N+1)), np.zeros((N+1,N+1)) # response matrix

for idx1,j in enumerate(smile_level):
    smile_vol_zap = sum(zap>j,1).astype(float)/data_dim
    smile_vol_non = sum(nonzap>j,1).astype(float)/data_dim
    for idx2,p in enumerate(perc):
        m_zap[idx1][idx2] = sum(smile_vol_zap>p).astype(float)/zap_num
        m_non[idx1][idx2] = sum(smile_vol_non>p).astype(float)/nonzap_num
```

```python
# plot volume effect
def show_smile_vol(m):
    imshow(m)
    xlabel('Portion of ad, (x)')
    ylabel('Smile response, (y)')
    colorbar()

fig = figure(figsize=(17,6))
subplot(1,2,1), show_smile_vol(m_zap)
subplot(1,2,2), show_smile_vol(m_non)
```
